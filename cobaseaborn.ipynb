```python
# Renaming 'price_detail__amount' to 'price'
df = df.rename(columns={'price_detail__amount': 'price'})
```

```python
# Distribution of course prices
sns.displot(df['price'], kde=True)
plt.show()
```

```python
# Checking for outliers in price column
sns.boxplot(x=df['price'])
plt.show()
```

```python
# Distribution of number of subscribers
sns.displot(df['num_subscribers'], kde=True)
plt.show()
```

```python
# Distribution of average ratings
sns.displot(df['avg_rating'], kde=True)
plt.show()
```

```python
# Distribution of Recent Average Ratings
sns.displot(df['avg_rating_recent'], kde=True)
plt.show()
```
